```python
import pygame
import random
import sys

# Inisialisasi pygame
pygame.init()

# Konstanta layar
SCREEN_WIDTH = 600
SCREEN_HEIGHT = 400
BALL_RADIUS = 25

# Membuat layar
screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("Contoh Pukul Bola - Anaconda Python")

# Posisi awal bola
ball_x = random.randint(BALL_RADIUS, SCREEN_WIDTH - BALL_RADIUS)
ball_y = random.randint(BALL_RADIUS, SCREEN_HEIGHT - BALL_RADIUS)

# Kecepatan bola
ball_speed_x = random.choice([-3, -2, 2, 3])
ball_speed_y = random.choice([-3, -2, 2, 3])

# Skor
score = 0

# Font
font = pygame.font.Font(None, 36)

# Clock
clock = pygame.time.Clock()

# Loop utama
running = True

while running:
    # Mengecek event
    for event in pygame.event.get():
        # Jika tombol close ditekan
        if event.type == pygame.QUIT:
            running = False
        
        # Jika mouse diklik
        if event.type == pygame.MOUSEBUTTONDOWN:
            # Ambil posisi mouse
            mx, my = pygame.mouse.get_pos()
            
            # Hitung jarak mouse ke bola ( rumus jarak Euclidean )
            distance = ((mx - ball_x) ** 2 + (my - ball_y) ** 2) ** 0.5
            
            # Jika bola terkena klik
            if distance <= BALL_RADIUS:
                score += 1
                
                # Posisi bola baru
                ball_x = random.randint(BALL_RADIUS, SCREEN_WIDTH - BALL_RADIUS)
                ball_y = random.randint(BALL_RADIUS, SCREEN_HEIGHT - BALL_RADIUS)
                
                # Kecepatan baru
                ball_speed_x = random.choice([-3, -2, 2, 3])
                ball_speed_y = random.choice([-3, -2, 2, 3])

    # Gerakkan bola
    ball_x += ball_speed_x
    ball_y += ball_speed_y

    # Pantulan kiri kanan
    if ball_x <= BALL_RADIUS or ball_x >= SCREEN_WIDTH - BALL_RADIUS:
        ball_speed_x = -ball_speed_x

    # Pantulan atas bawah
    if ball_y <= BALL_RADIUS or ball_y >= SCREEN_HEIGHT - BALL_RADIUS:
        ball_speed_y = -ball_speed_y

    # Warna background (putih)
    screen.fill((255, 255, 255))

    # Gambar bola (merah)
    pygame.draw.circle(
        screen,
        (255, 0, 0),
        (int(ball_x), int(ball_y)),
        BALL_RADIUS
    )

    # Tampilkan skor
    text = font.render(f"Score: {score}", True, (0, 0, 0))
    screen.blit(text, (10, 10))

    # Update layar
    pygame.display.update()

    # FPS (60 frame per detik)
    clock.tick(60)

# Keluar dari pygame dengan bersih
pygame.quit()
sys.exit()
```
