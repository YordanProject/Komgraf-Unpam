```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load Gambar
gambar = cv2.imread('gambar.jpg', cv2.IMREAD_GRAYSCALE)

# Hitung histogram
histogram = cv2.calcHist([gambar], [0], None, [256], [0, 256])

# Tampilkan histogram
plt.figure(figsize=(7, 5))
plt.title('Gambar Histogram')
plt.xlabel('Intensitas Pixel')
plt.ylabel('Jumlah pixel')
plt.plot(histogram, color='steelblue')
plt.xlim(0, 256)
plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load gambar
img = cv2.imread('gambar.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel_size = (5, 5)
mean_filtered = cv2.blur(img_rgb, kernel_size)

# Hasil Perbandingan
plt.figure(figsize=(10, 5))

# Gambar Asli
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title('Gambar Asli')
plt.axis('off')

# Hasil Mean Filter
plt.subplot(1, 2, 2)
plt.imshow(mean_filtered)
plt.title(f'Mean Filter {kernel_size}')
plt.axis('off')

plt.tight_layout()
plt.show()
```
