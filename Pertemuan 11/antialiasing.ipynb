```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 1000)
y = np.sin(x)

# Dengan AA
plt.plot(x, y, label='Dengan AA', antialiased=True)
plt.plot(x, 1.5 * y + 0.5, label='Tanpa AA', antialiased=False)
plt.title('Perbandingan Antialiasing')
plt.legend(loc='lower left')

plt.show()
```
