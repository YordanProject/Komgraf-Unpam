```python
import numpy as np
import matplotlib.pyplot as plt

# Segitiga awal
pts = np.array([[0,0], [2,0], [1,2], [0,0]])

# Hasil Shear (k=0.8)
pts_h = np.column_stack([pts[:,0] + 0.8*pts[:,1], pts[:,1]])

# Plot
plt.figure(figsize=(10,4))

plt.subplot(121)
plt.plot(pts[:,0], pts[:,1], 'b-o')
plt.title('Asli')
plt.axis('equal')

plt.subplot(122)
plt.plot(pts_h[:,0], pts_h[:,1], 'r-o')
plt.title('Shear (k=0.8)')
plt.axis('equal')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def rotasi(x, y, sudut):
    rad = np.radians(sudut)
    x_baru = x * np.cos(rad) - y * np.sin(rad)
    y_baru = x * np.sin(rad) + y * np.cos(rad)
    return x_baru, y_baru

shape = np.array([[1, 1], [3, 1], [3, 3], [1, 3], [1, 1]])

shape_rotasi = np.array([rotasi(x, y, 45) for x, y in shape])

plt.figure(figsize=(6,6))

plt.grid(True, linestyle='-', alpha=0.7)

plt.axhline(0, color='black', linewidth=1.2)
plt.axvline(0, color='black', linewidth=1.2)

plt.plot(shape[:,0], shape[:, 1], 'b-', label='Asli')
plt.plot(shape_rotasi[:,0], shape_rotasi[:, 1], 'r-', label='Rotasi 45°')

plt.xlim(-2, 4)
plt.ylim(-1, 5)

plt.axis('equal')
plt.legend()
plt.show()
```
