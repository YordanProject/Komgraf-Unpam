```python
import matplotlib.pyplot as plt
import numpy as np

def dda_line(x1, y1, x2, y2):
    
    dx = x2 - x1
    dy = y2 - y1
    
    steps = max(abs(dx), abs(dy))
    
    x_inc = dx / steps
    y_inc = dy / steps
    
    x, y = x1, y1
    points_x, points_y = [round(x)], [round(y)]
    
    for i in range(int(steps)):
        x += x_inc
        y += y_inc
        points_x.append(round(x))
        points_y.append(round(y))
        
    return points_x, points_y

x_points, y_points = dda_line(2, 3, 10, 8)

plt.plot(x_points, y_points, 'bo-') 
plt.title("Algoritma DDA - Garis dari (2,3) ke (10,8)")
plt.grid(True)
plt.show()
```

```python
def bresenham_line(x1, y1, x2, y2):
    points_x, points_y = [], []
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    sx = 1 if x2 > x1 else -1
    sy = 1 if y2 > y1 else -1
    x, y = x1, y1
    
    if dx > dy:
        p = 2 * dy - dx
        for _ in range(dx + 1):
            points_x.append(x)
            points_y.append(y)
            x += sx
            if p < 0: p += 2 * dy
            else:
                y += sy
                p += 2 * (dy - dx)
    else:
        p = 2 * dx - dy
        for _ in range(dy + 1):
            points_x.append(x)
            points_y.append(y)
            y += sy
            if p < 0: p += 2 * dx
            else:
                x += sx
                p += 2 * (dx - dy)
    return points_x, points_y

xp, yp = bresenham_line(2, 3, 10, 8)
plt.plot(xp, yp, 'ro-') 
plt.title("Algoritma Bresenham")
plt.grid(True)
plt.show()
```
