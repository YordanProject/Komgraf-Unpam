```python
from PIL import Image
import matplotlib.pyplot as plt

# Load Image
img = Image.open('yordan.jpeg')
w, h = img.size

# Definisikan area crop
crop_areas = {
    "Atas Kiri": (0, 0, int(w*0.5), int(h*0.5)),
    "Tengah": (int(w*0.25), int(h*0.25), int(w*0.75), int(h*0.75)),
    "Bawah Kanan": (int(w*0.110), int(h*0.5), w, h)
}

# Tampilkan hasil crop
plt.figure(figsize=(15, 10))

# Gambar asli
plt.subplot(1, 4, 1)
plt.imshow(img)
plt.title("Asli")
plt.axis('off')

# Hasil
for i, (nama, area) in enumerate(crop_areas.items(), 2):
    cropped = img.crop(area)

    plt.subplot(1, 4, i)
    plt.imshow(cropped)
    plt.title(nama)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
