```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

theta = np.linspace(0, 2*np.pi, 100)
r = 1

# Lingkaran 1: bidang XY (z = 0)
x1 = r * np.cos(theta)
y1 = r * np.sin(theta)
z1 = np.zeros_like(theta)
ax.plot(x1, y1, z1, 'b-', linewidth=2, label='Lingkaran di XY plane')

# Lingkaran 2: bidang XZ (y = 0)
x2 = r * np.cos(theta)
y2 = np.zeros_like(theta)
z2 = r * np.sin(theta)
ax.plot(x2, y2, z2, 'r-', linewidth=2, label='Lingkaran di XZ plane')

# Lingkaran 3: bidang YZ (x = 0)  ← INI YANG SALAH DI KODE ASLI
x3 = np.zeros_like(theta)
y3 = r * np.cos(theta)
z3 = r * np.sin(theta)
ax.plot(x3, y3, z3, 'g-', linewidth=2, label='Lingkaran di YZ plane')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Lingkaran dalam 3 Dimensi - YordanKece')
ax.legend()
plt.show()
```
