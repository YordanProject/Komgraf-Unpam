```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load gambar
img = cv2.imread('gambar.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Kernel
kernel = np.ones((3, 3), dtype=np.uint8)

# Dilasi
dilasi = cv2.dilate(img_gray, kernel, iterations=1)

# Hasil
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('Gambar Asli')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(dilasi, cmap='gray')
plt.title('Hasil Dilasi')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load
img = cv2.imread('gambar.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Scaling
height, width = img_rgb.shape[:2]

# Fixed Size (100x50)
img_fixed = cv2.resize(img_rgb, (100, 50), interpolation=cv2.INTER_LINEAR)

# Skala Rasio 2x
img_scaled_ratio = cv2.resize(img_rgb, (width * 2, height * 2),
                               interpolation=cv2.INTER_LINEAR)

# Print ukuran
print(f"Ukuran Asli              : {img_rgb.shape[:2]}")
print(f"Ukuran Fixed Size        : {img_fixed.shape[:2]}")
print(f"Ukuran Rasio (2x Lipat)  : {img_scaled_ratio.shape[:2]}")

# Hasil Perbandingan
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(img_rgb)
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_fixed)
plt.title('Fixed Size (100x50)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img_scaled_ratio)
plt.title('Skala Rasio 2x')
plt.axis('off')

plt.tight_layout()
plt.show()
```
